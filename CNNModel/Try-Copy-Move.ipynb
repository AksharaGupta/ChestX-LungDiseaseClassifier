```python
import shutil
import pandas as pd
import numpy as np
CSV2 = pd.read_csv("sample_labels.csv")
```

```python
CSV = CSV2[['Image Index','Finding Labels']]
CSV.head(10)
```

```python
#CSV.iloc[0,1]
```

```python
for i in range(5605):
    CSV.iloc[i,1] = CSV.iloc[i,1].split("|")
```

```python
CSV.head(20)
```

```python
#Run this cell only if the folders do not exist in the below mentioned directory
import os
diseases = ['Cardiomegaly','Emphysema','Effusion','Hernia','Nodule','Pneumothorax','Atelectasis','Pleural_Thickening','Mass','Edema','Consolidation','Infiltration','Fibrosis','Pneumonia','No Finding']
for d in diseases:
    os.mkdir('Using Keras/data/train/'+d)
    os.mkdir('Using Keras/data/test/'+d)
```

```python
for i in range(3500):
    src = "sample/images/" + CSV.iloc[i,0]
    for j in CSV.iloc[i,1]:
        dst = "Using Keras/data/train/" + j + "/" + CSV.iloc[i,0]  #j is each label for an image
        shutil.copy2(src,dst)
```

```python
for i in range(3500,5000):
    src = "sample/images/" + CSV.iloc[i,0]
    for j in CSV.iloc[i,1]:
        dst = "Using Keras/data/test/" + j + "/" + CSV.iloc[i,0]
        shutil.copy2(src,dst)
```
